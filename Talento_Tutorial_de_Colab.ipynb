Para ***ejecutar*** o ***correr*** una celda de código, basta poner el cursor en la celda y tocar el botón de "play" (ver imagen abajo) o presionar ***shift*** + ***enter*** en el teclado. Ejectudada la celda, aparecerá debajo el "output" o lo que devuelve el código que corrimos.
